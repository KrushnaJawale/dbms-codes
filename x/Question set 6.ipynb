#### Question set 6

#### Write a Python program to develop a regression model using linear regression to predict house prices and evaluate its performance using mean squared error.

#### Step 1: Import libraries

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

#### Step 2: Sample dataset (Area vs Price)

```python
# Area (sq ft)
X = np.array([500, 800, 1000, 1200, 1500]).reshape(-1, 1)
```

```python
X
```

```python
# Price (in lakhs)
y = np.array([50, 80, 100, 120, 150])
```

#### Step 3: Split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#### Step 4: Create model

```python
model = LinearRegression()
```

#### Step 5: Train model

```python
model.fit(X_train, y_train)
```

#### Step 6: Predict

```python
y_pred = model.predict(X_test)
```

#### Step 7: Evaluate model

```python
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
r2_score(y_test, y_pred)
```

#### Step 8: Predict new house price

```python
new_area = [[1200]]
predicted_price = model.predict(new_area)
print("Predicted Price for 1200 sq ft:", predicted_price[0])
```

#### Viva Questions & Answers

**1. What is Linear Regression?**
- Linear Regression is a supervised learning algorithm used to predict continuous values using a straight-line relationship between variables.

**2. What is the dependent and independent variable in this model?**
- Independent variable: Area (input)
- Dependent variable: House price (output)

**3. What does the model learn during training?**
- It learns the best-fit line (relationship) between area and price.

**4. What is Mean Squared Error (MSE)?**
- MSE is the average of squared differences between actual and predicted values; lower MSE means better performance.

**5. How can you predict the price of a new house?**
- By giving a new input (area) to the trained model using model.predict().
