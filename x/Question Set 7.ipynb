**Question Set 7**

**Implement a Markov Chain model using Python to simulate machine state transitions for given steps.**

#### Step 1: Define Machine States

```python
states = ["Operational", "Minor Wear", "Major Fault", "Failure"]
```

#### Step 2: Define Transition Probability Matrix

This matrix represents the probability of moving from one state to another:

```python
import numpy as np

P = np.array([
    [0.7, 0.2, 0.1, 0.0],
    [0.2, 0.5, 0.2, 0.1],
    [0.0, 0.3, 0.4, 0.3],
    [0.0, 0.0, 0.0, 1.0]
])
```

#### Step 3: Initialize System

Start from an initial state (e.g., Operational):

```python
state = 0
history = []
```

#### Step 4: Simulate State Transitions

Simulate machine behavior over time:

```python
for i in range(15):
    history.append(state)
    print("Step", i+1, ":", states[state])
    state = np.random.choice([0,1,2,3], p=P[state])
```

#### Step 5: Visualize Results

Plot the machine state transitions:

```python
import matplotlib.pyplot as plt

plt.plot(history, marker='o')
plt.title("Machine State Transition Over Time")
plt.xlabel("Time Step")
plt.ylabel("State")
plt.yticks([0,1,2,3], states)
plt.grid()
plt.show()
```

#### Step 6: Analysis

- If the machine frequently reaches Major Fault (S3) → maintenance is required.
- If it reaches Failure (S4) → immediate repair is needed.
- Helps in planning predictive maintenance.

**1. What is a Markov Process?**

A Markov Process is a stochastic process where the next state depends only on the current state, not on past states.

**2. What is a Markov Chain?**

A Markov Chain is a mathematical model that describes transitions between different states with fixed probabilities.

**3. What are the machine states used in this experiment?**

Operational, Minor Wear, Major Fault, and Failure.

**4. What is a transition probability matrix?**

It is a matrix that shows the probability of moving from one state to another in a Markov process.

**5. Why is the Failure state considered absorbing?**

Because once the machine reaches failure, it remains in that state (probability = 1).
