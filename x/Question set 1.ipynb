### Question set 1

#### Write a Python program to classify emails as spam or not spam using a simple dataset with columns label and message.

```python
data  = {
    'label': ['spam', 'ham', 'spam', 'ham', 'spam'],
    'message': [
        'Win money now',
        'Hello how are you',
        'Claim your prize',
        'Let us meet tomorrow',
        'Congratulations you won',
    ]
}
```

#### Step 1: Import libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

#### Step 2: Sample dataset

```python
data = pd.DataFrame({
    'label': ['spam', 'ham', 'spam', 'ham', 'spam'],
    'message': [
        'Win money now',
        'Hello how are you',
        'Claim your prize',
        'Let us meet tomorrow',
        'Congratulations you won',
    ]
})
```

```python
data
```

#### Step 3: Convert labels (spam=1, ham=0)

```python
data['label'] = data['label'].map({'spam':1, 'ham':0})
```

```python
data
```

#### Step 4: Convert text to numerical features

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['message'])
y = data['label']
```

```python
import numpy as np
```

#### Step 5: Split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#### Step 6: Train model

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

#### Step 7: Predict

```python
y_pred = model.predict(X_test)
y_pred
```

#### Step 8: Evaluate

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

#### Step 9: Test with new email

```python
new_msg = ["Free prize available"]
new_X = vectorizer.transform(new_msg)
prediction = model.predict(new_X)

print("Spam" if prediction[0] == 1 else "Not Spam")
```
