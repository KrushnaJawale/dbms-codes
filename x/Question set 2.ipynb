### Question Set 2

#### Write a Python program to read a vibration signal from a given data (mm column) and plot its graph with respect to time. Also, find and display the dominant frequency using FFT.

#### Given 

```python
data = {"mm": [-1.03, 0.65, -0.58, -0.53, 0.97, -0.51, -0.80, 0.71, -0.80, -0.76]}
```

#### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Load data

```python
data
```

```python
data = pd.DataFrame(data)
signal = data['mm']
```

#### Create time axis (assume sampling interval = 0.01 sec)

```python
dt = 0.01
t = np.arange(0, len(signal)*dt, dt)
```

```python
t
```

#### Time-domain plot

```python
plt.plot(t, signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Time Domain Signal')
plt.grid()
plt.show()
```

### FFT

```python
f = np.fft.fftfreq(len(signal), d=dt)
fft_vals = np.abs(np.fft.fft(signal))
```

```python
fft_vals
```

#### Positive frequencies

```python
10/2
```

```python
f = f[:len(f)//2]
fft_vals = fft_vals[:len(f)]
```

```python
fft_vals
```

#### Frequency-domain plot

```python
plt.plot(f, fft_vals)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Frequency Domain')
plt.grid()
plt.show()
```

#### Dominant frequency

```python
print("Dominant Frequency:", f[np.argmax(fft_vals)], "Hz")
```

#### Viva Revision Sheet

**1. What is a vibration signal?**
- It is a signal that represents oscillations or movements of a system over time.

**2. What is the difference between time-domain and frequency-domain analysis?**
- Time-domain shows how the signal changes with time, while frequency-domain shows the different frequencies present in the signal.

**3. What is FFT and why is it used?**
- FFT (Fast Fourier Transform) is used to convert a signal from time domain to frequency domain for analysis.

**4. What is meant by dominant frequency?**
- It is the frequency with the highest magnitude in the signal.

**5. Why do we consider only positive frequencies in FFT?**
- Because FFT output is symmetric, and positive frequencies contain all useful information.
