### Question set 4

#### Write a Python program to implement handwritten digit recognition using the MNIST dataset. Train a simple neural network model and display the prediction of a test image.

#### Step 1: Import libraries

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

#### Step 2: Load MNIST dataset

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

#### Step 3: Normalize data (0–255 → 0–1)

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

#### Step 4: Build simple model

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

#### Step 5: Compile model

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

#### Step 6: Train model

```python
model.fit(x_train, y_train, epochs=3)
```

#### Step 7: Evaluate model

```python
loss, acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", acc)
```

#### Step 8: Predict one image

```python
prediction = model.predict(x_test)
```

#### Show image

```python
plt.imshow(x_test[10], cmap='gray')
plt.title("Predicted Digit: " + str(prediction[10].argmax()))
plt.show()
```

#### Viva Questions & Answers

**1. What is the MNIST dataset?**
- MNIST is a dataset of 28×28 pixel grayscale images of handwritten digits (0–9) used for training image classification models.

**2. Why do we normalize the image data?**
- To scale pixel values from 0–255 to 0–1, which improves model training and accuracy.

**3. What is the role of the Flatten layer?**
- It converts a 2D image (28×28) into a 1D array so it can be fed into the neural network.

**4. Why is softmax activation used in the output layer?**
- Softmax converts outputs into probabilities for each digit (0–9), helping in classification.

**5. What does accuracy represent in this model?**
- Accuracy shows the percentage of correctly predicted digits out of total predictions.
