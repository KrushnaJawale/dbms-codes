### Question set 3

#### Write a Python program to read a dataset of news articles and extract the word count of each article as a feature for training data.

```python
data = {"title": ["Self-Organization of Associative Database", "Boosting Decision Trees"
          ,"Family Discovery", "Neural Control for Nonlinear Dynamic Systems", 
                  "A Model of Auditory Streaming"] }
```

#### Step 1: Import libraries

```python
import pandas as pd
```

#### Step 2: Load dataset

```python
df = pd.DataFrame(data)
df
```

#### Step 3: Display data

```python
print(df.head())   #5 records
```

#### Step 3: Extract article column (assume first column)

```python
articles = df.iloc[:, 0]  #[Rows, columns]
articles
```

#### Step 4: Feature extraction - word count

```python
df['Word_Count'] = articles.apply(lambda x: len(str(x).split()))
```

#### Step 5: Display result

```python
print("\nData with Word Count:\n", df.head())
```

#### Step 6: Prepare training data

```python
X = df['Word_Count']   # Feature
print("\nTraining Feature (Word Count):\n", X)
```
